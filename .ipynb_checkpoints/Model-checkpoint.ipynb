```python
# One hot encoding to change sex column 
train = pd.get_dummies(train, dtype=int, columns = ['Sex'])
train = train.rename(columns={'Sex_2': 'Female', 'Sex_0': 'Infant', 'Sex_1': 'Male'})
test = pd.get_dummies(test, dtype=int, columns = ['Sex'])
test = test.rename(columns={'Sex_2': 'Female', 'Sex_0': 'Infant', 'Sex_1': 'Male'})
```

```python
#Whole weight: Whole weight indicates the total weight of the abalone, encompassing both the meat and the shell, measured in grams.

#Whole weight.1 (Shucked weight): This feature represents the weight of the abalone meat only, measured in grams. It indicates the amount of meat extracted from the shell.

#Whole weight.2 (Viscera weight): Viscera weight signifies the weight of the abalone gut after bleeding, measured in grams.It provides insights into the weight of the internal organs of the abalone.

#Shell weight: This feature represents the weight of the abalone shell only, excluding the meat, measured in grams. It provides insights into the weight distribution between the shell and the edible portion of the abalone.

train = train.rename(columns={'Whole weight.1': 'Shucked weight', 'Whole weight.2': 'Viscera weight'})
train['Shell Ratio'] = train['Shell weight'] / train['Whole weight']
train['Shucked Ratio'] =  train['Shucked weight'] / train['Whole weight']
train['Viscera Ratio'] = train['Viscera weight'] / train['Whole weight']

test = test.rename(columns={'Whole weight.1': 'Shucked weight', 'Whole weight.2': 'Viscera weight'})
test['Shell Ratio'] = test['Shell weight'] / test['Whole weight']
test['Shucked Ratio'] =  test['Shucked weight'] / test['Whole weight']
test['Viscera Ratio'] = test['Viscera weight'] / test['Whole weight']
```

```python
# Volume: Combining length, diameter, and height
train['Volume'] = train['Length'] * train['Diameter'] * train['Height']

# Surface Area: Using length and diameter
train['Surface Area'] = np.pi * ((train['Length'] * train['Diameter']) + (train['Length'] * train['Height']) + (train['Diameter'] * train['Height']))

# Interaction Features: Whole weight with other features
train['Weight-Length'] = train['Whole weight'] * train['Length']
train['Weight-Diameter'] = train['Whole weight'] * train['Diameter']
train['Weight-Height'] = train['Whole weight'] * train['Height']

test['Volume'] = test['Length'] * test['Diameter'] * test['Height']
test['Surface Area'] = np.pi * ((test['Length'] * test['Diameter']) + (test['Length'] * test['Height']) + (test['Diameter'] * test['Height']))

test['Weight-Length'] = test['Whole weight'] * test['Length']
test['Weight-Diameter'] = test['Whole weight'] * test['Diameter']
test['Weight-Height'] = test['Whole weight'] * test['Height']
```

```python
def algo_test(x,y):
        L=LinearRegression()
        R=Ridge()
        #Lass=Lasso()
        #E=ElasticNet()
        sgd=SGDRegressor()
        #ETR=ExtraTreeRegressor()
        GBR=GradientBoostingRegressor()
        kn=KNeighborsRegressor()
        ada=AdaBoostRegressor()
        #dt=DecisionTreeRegressor()
        xgb=XGBRegressor()
        svr=SVR()
        mlp_regressor = MLPRegressor()

       
        
        algos=[L,R,sgd,GBR,ada,kn,xgb,svr,mlp_regressor]
        algo_names=['Linear','Ridge','SGD','Gradient Boosting',
                    'KNeighborsRegressor','AdaBoost','XGBRegressor','SVR','mlp_regressor']
        
        x_train, x_test, y_train, y_test=train_test_split(x,y,test_size = 0.20, random_state=42)
        
        r_squared= []
        rmse= []
        rmsle= []
        mae= []
        
        # mean absolute error
        #r^2 is goodness of fit of model
        # Add RMSLE for outliers as it works better when there are outliers ????
        result=pd.DataFrame(columns=['R_Squared','RMSE', 'MAE'],index=algo_names)
        
        
        for algo in algos:
            p=algo.fit(x_train,y_train).predict(x_test)
            r_squared.append(r2_score(y_test,p))
            rmse.append(mean_squared_error(y_test,p)**.5)
            #rmsle.append(mean_squared_log_error(y_test, p, squared=False))
            mae.append(mean_absolute_error(y_test,p))
        
            

        result.R_Squared=r_squared
        result.RMSE=rmse
        result.MAE=mae
        
        rtable=result.sort_values('R_Squared',ascending=False)
        return rtable
```
